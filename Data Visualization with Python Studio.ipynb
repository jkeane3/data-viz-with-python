```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.info()
df.head(10)
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.

df['year'].value_counts()[:20].plot(kind='bar', color='blue')
plt.title('Books Published by Year')
plt.xlabel('Year')
plt.ylabel('Books Published')

plt.show()
```

```python
df['publisher'].value_counts()[:20].plot(kind='barh', color='blue')

plt.title('Books Published by Publisher')
plt.xlabel('Books Publsihed')
plt.ylabel('Publisher')

plt.show()
```

```python
# Use your Python skills to create a visualization here.

#Creating the scatter plot
plt.scatter(df['ratings_count'],df['text_reviews_count'],alpha=0.5 )
#Adding the aesthetics
plt.title('Ratings count vs Text Reviews count')
plt.xlabel('Ratings count')
plt.ylabel('Text Reviews count') 
#Show the plot
plt.show()
```

```python
sb.scatterplot('text_reviews_count','ratings_count', data = df, s = 200, linewidth = 3, edgecolor = 'black',color="lightgreen")
plt.title('Text ratings count versus reviews count', color = 'black', fontsize = 18)
plt.savefig('scatter_review.png')

plt.show()
```

```python
sb.regplot(x = 'text_reviews_count', y = 'ratings_count', data = df, color = 'green', scatter_kws = {'s':150,'alpha':0.4})
plt.title('Text ratings count versus reviews count', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('text_reviews_count', color = 'black')
plt.ylabel('ratings_count', color = 'black')
plt.savefig('reg_review_ratings.png')

plt.show()
```
